```python
import numpy as np
import matplotlib.pyplot as plt

n = 200  # or 500, must be even
t = np.arange(n)

# X_{0t}
Z0 = np.random.normal(0, 1)
X0t = np.full(n, Z0)

# X_{jt} for 1 <= j < n/2
Xjt_sum = np.zeros(n)
for j in range(1, n//2):
    omega_j = 2 * np.pi * j / n
    Zj1 = np.random.normal(0, np.sqrt(2))
    Zj2 = np.random.normal(0, np.sqrt(2))
    Xjt = Zj1 * np.cos(omega_j * t) + Zj2 * np.sin(omega_j * t)
    Xjt_sum += Xjt

# X_{n/2, t}
Zn2 = np.random.normal(0, 1)
Xn2t = Zn2 * (-1)**t

# Final sum
X = (X0t + Xjt_sum + Xn2t) / np.sqrt(n)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, X, label='Special $X_t$')
plt.xlabel('Time')
plt.ylabel('$X_t$')
plt.title('Simulated Special Time Series')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Simulate IID N(0,1)
X_iid = np.random.normal(0, 1, n)

# Plot the IID series
plt.figure(figsize=(10, 4))
plt.plot(t, X_iid, label='IID $N(0,1)$')
plt.xlabel('Time')
plt.ylabel('$X_t$')
plt.title('IID $N(0,1)$ Time Series')
plt.legend()
plt.tight_layout()
plt.show()
```
